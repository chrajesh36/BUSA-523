Write a program for “Rock-Paper-Scissors-Lizard-Spock”.

```python
import random
randomValue=random.randint(1,5)
if randomValue==1:
    computer="Rock"
elif randomValue==2:
    computer="Paper"
elif randomValue==3:
    computer="Scissors"
elif randomValue==4:
    computer="Spoke"
else:
    computer="Lizard"    
user=input("Enter your move Rock, Paper, Scissors, Spoke or Lizard : ")
n=1
Game_Result=0
while n!=0:
    if user == "Rock" or user == "Paper" or user == "Scissors" or user == "Spoke" or user == "Lizard":
        break
    else:
        user=input(f"Wrong input, Retry with one option of these: Rock, Paper, Scissors, Spoke or Lizard : ")
    n+=1
    if n==5:
        Game_Result="NA"
        break
if Game_Result=="NA":
    print("Too Many wrong inputs try new game. Please enter correct input next time.")
if Game_Result!="NA":
    print(f"Computer's option is: '{computer}'.")
if (user==computer):
    Game_Result="tie"
if (user=="Rock" and computer=="Paper") or (user=="Paper" and computer=="Rock"):
    print("Result is : Paper covers Rock.")
    if user=="Paper":
        Game_Result="win"
    else:
        Game_Result="loss"
if (user=="Rock" and computer=="Scissors") or (user=="Scissors" and computer=="Rock"):
    print("Result is : Rock crushes Scissors.")
    if user=="Rock":
        Game_Result="win"
    else:
        Game_Result="loss"     
if (user=="Rock" and computer=="Spoke") or (user=="Spoke" and computer=="Rock"):
    print("Result is : Spoke vaporizes Rock.")
    if user=="Rock":
        Game_Result="loss"
    else:
        Game_Result="win" 
if (user=="Rock" and computer=="Lizard") or (user=="Lizard" and computer=="Rock"):
    print("Result is : Rock crushes Lizard.")
    if user=="Rock":
        Game_Result="win"
    else:
        Game_Result="loss"
if (user=="Paper" and computer=="Scissors") or (user=="Scissors" and computer=="Paper"):
    print("Result is : Scissors cuts Paper.")
    if user=="Scissors":
        Game_Result="win"
    else:
        Game_Result="loss"
if (user=="Paper" and computer=="Spoke") or (user=="Spoke" and computer=="Paper"):
    print("Result is : Paper disproves Spoke.")
    if user=="Paper":
        Game_Result="win"
    else:
        Game_Result="loss"
if (user=="Paper" and computer=="Lizard") or (user=="Lizard" and computer=="Paper"):
    print("Result is : Lizard eats Paper.")
    if user=="Lizard":
        Game_Result="win"
    else:
        Game_Result="loss"
if (user=="Scissors" and computer=="Spoke") or (user=="Spoke" and computer=="Scissors"):
    print("Result is : Spoke smashes Scissors.")
    if user=="Spoke":
        Game_Result="win"
    else:
        Game_Result="loss"         
if (user=="Scissors" and computer=="Lizard") or (user=="Lizard" and computer=="Scissors"):
    print("Result is : Scissors decapitates Lizard.")
    if user=="Scissors":
        Game_Result="win"
    else:
        Game_Result="loss"
if (user=="Spoke" and computer=="Lizard") or (user=="Lizard" and computer=="Spoke"):
    print("Result is : Lizard poisions Spoke.")
    if user=="Lizard":
        Game_Result="win"
    else:
        Game_Result="loss"     
if Game_Result=="win":
    print(f"Congratulations you won the game.")
if Game_Result=="loss":
    print(f"Sorry you lost the game, better luck next time.")
if Game_Result=="tie":
    print(f"Its a tie, both selected same option '{user}'.")    
```
